# Correlation Analysis

```python
#budget high correlation
#company high correlation
```

```python
#scatter plot of budget vs gross
plt.scatter(x = df['budget'], y = df['gross'])
plt.title("Budget vs Gross Earnings")
plt.ylabel("Budget for film")
plt.xlabel("Gross Earnings")
plt.show();
```

```python
df[['budget','gross']].head()
```

```python
#Regression plot - plot budget vs gross using seaborn

sns.regplot(
            x = 'budget', y = 'gross', data = df, 
            scatter_kws = {"color": "red"},
            line_kws = {"color": "blue"}
           );
```

```python
#determinig the actual correlation - pearson, kendall, spearman

df.corr(method='pearson')
```

```python
#high correlation between budget for films and gross earnings
```

```python
corr_matrix = df.corr(method='pearson')

sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation Matrix for Numeric features")
plt.ylabel("Movie Features")
plt.xlabel("Movie Features")
plt.show();
```

```python
#Looking at company

df.nunique()
```

```python
df_numerized = df[:]

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(2)
```

```python
df.head(2)
```

```python
corr_matrix = df_numerized.corr(method='pearson')

sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation Matrix for Numeric features")
plt.ylabel("Movie Features")
plt.xlabel("Movie Features")
plt.show();
```

```python
#Correlation Unstacking

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
```

```python
pd.reset_option('^display.', silent=True)
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
